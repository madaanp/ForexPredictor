```python
df.columns = map(str.upper, df.columns)
df.rename(columns=lambda x:x+'_USD', inplace=True)
df.rename(columns={'TIME SERIES_USD':'Time Series'}, inplace=True)
df.shape
```

```python
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['Time Series'] = pd.to_datetime(df['Time Series'])
df['month'] = df['Time Series'].dt.month
df['year'] = df['Time Series'].dt.year
df['month_year'] = df['Time Series'].dt.to_period('M')
```

```python
from matplotlib import pyplot as plt

x = df['month_year'].astype(str)
y = df['EURO_USD']

plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.title("Exchange Rate: EURO/USD")
plt.xlabel("Month")
plt.ylabel("Exchange Rate")
plt.show()
```

```python
groupby_all = df.groupby('month_year').apply(lambda x: pd.Series({
                                                    'AUD_mean': x['AUSTRALIAN DOLLAR_USD'].mean(),
                                                    'EUR_mean': x['EURO_USD'].mean(),
                                                    'NZD_mean': x['NEW ZEALAND DOLLAR_USD'].mean(),
                                                    'SGD_mean': x['SINGAPORE DOLLAR_USD'].mean(),
                                                    'GBP_mean': x['U.K. POUND_USD'].mean()
})).reset_index()

groupby_all = pd.DataFrame(groupby_all)

groupby_all['USD_mean'] = [1 for i in range(119)]
groupby_all.head()
```
